<h1 style="text-decoration: underline;"> Exercice 1 : Cours de Pétrole </h1>

```python
import openpyxl
import pandas as pd
import matplotlib.pyplot as plt

wb = openpyxl.load_workbook('petrole.xlsx','r','UTF-8')
fichier = wb.sheetnames
ws = wb[fichier[0]]
D=[];P=[]
for ligne in ws.iter_rows(min_row=5, values_only=True):
    D.append(pd.to_datetime(ligne[0]))
    P.append(float(ligne[1]))
s = pd.Series(P, index=D)
print('étude statistique :')
print(s.describe())
print('jour du prix le plus faible =', s.idxmin())
print('jour du prix le plus élevé =', s.idxmax())

plt.figure(figsize=(20,10))
plt.subplot(1, 3, 1)
s.plot()
plt.title('Etude graphique')
plt.xlabel('Date')
plt.ylabel('Prix')
plt.grid(True)

plt.subplot(1, 3, 3)
s.hist()
plt.title('Etude graphique')
plt.grid(True)
plt.show()
```

<h1 style="text-decoration: underline;"> Exercice 3 : Série statistique </h1>

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

pokemon = pd.read_excel("ventes.xlsx",dtype ={'Prix':np.float64} , engine='openpyxl')
dff = pd.DataFrame()
X=pokemon.groupby('Produit').mean()['Prix']
#print(X)
plt.subplot(1, 3, 1)
plt.title('Evolution du prix moyen de chaque carte')
X.plot()

plt.subplot(1, 3, 3)
plt.title('Répartition du prix des cartes')
X.hist()
plt.show()

print('moyenne = ', X.mean())
print('écarte-type = ', X.std())
print('carte la plus chère = ', X.idxmax()) #affiche la valeur qui se trouve dans la colonne par laquelle on a 
                                            #relié notre DataFrame (=produit) pour la valeur Prix max
print('carte la moins chère = ', X.idxmin()) #Idem avec le prix le plus faible
```

<h2><li> Question 3) </li></h2>

```python
V=dfff['Prix']['2022']
M=V.resample('M').mean()
W=V.resample('W').mean()
M.plot()
W.plot()
plt.title('Evolution des prix par semaine et par mois')
plt.xlabel('date')
plt.ylabel('prix')
plt.grid(True)
plt.show()
```
