```python
# Parser la date de la chaine de type: Vendu 25 sept. 2021

#L'idée était de créer un dictionnaire pour faire correspondre les mois de l'année en 'chiffre' et en 'lettre' càd janv==1,fev==2...
#Ainsi, on fait tourner la chaine de caractère dans la fonction jour pour obtenir "9"→"09"
#On a plus qu'a afficher l'année, le jour et le mois 
import numpy as np
#jour=str(input("jour="))
#mois=str(input("mois="))
#annee=str(input("année="))
L=['Vendu',jour,mois,annee]

n=L[1]
mois_saisie=L[2]
p = L[3]

ll={'1':['Janvier'],'2':['Fevrier'],'3':['Mars'],'4':['avril'],'5':['mai'],'6':['juin'],'7':['juillet'],'8':['aout'],'9':['Septembre'],'10':['octobre'],'11':['novembre'],'12':['decembre']}

def mois(dic,mois_saisie):
    key=''
    for (k,v) in dic.items():
        if mois_saisie==v: #Cette ligne est probablement l'erreur // l'idée est de comparer les valeurs à la saisie et de retourner la clé 
                 #pour laquelle la valeur et la saisie sont égales
            key=k
    return key

def jour(n):
    if int(n)<10:
        return "0"+str(n)
    else:
        return str(n)

print(p, end="-")
print(jour(n), end='-')
print(jour(mois(ll)), end="-")

```
