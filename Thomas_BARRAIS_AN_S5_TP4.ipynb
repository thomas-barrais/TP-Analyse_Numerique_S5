```python
from bs4 import BeautifulSoup
from openpyxl import Workbook
import urllib.request
# cr√©e le workbook
wb=Workbook()

# saisir la feuille de calcul active
ws = wb.active

# Les donn√©es peuvent √™tre affect√©es directement aux cellules
ws['A1']= "Titre"
ws['B1'] = "Prix"
ws['C1'] = "Date"

url="https://www.ebay.fr/sch/i_html?_from=R40&_nkw=Pikachu+58%2F102&_sacat=0&rt=nc&LH_Sold=1&LH_Complete=1&_ipg=200"
f=urllib.request.urlopen(url)
html_doc = f.read()
soup = BeautifulSoup(html_doc,"html.parser")
L=soup.select("li.s-item")[1:]
for k in L:
    titre=k.select("div.s-item__title")[0].text
    prix=k.select(".s-item__price .POSITIVE")[0].text
    date = k.select(".s-item__title--tagblock .POSITIVE")[0].text
    ws.append([1*titre, 1*prix, 1*date]) #on ajoute les items dans le workbook  
    print(" Titre=",titre,"\n Prix =",prix,"\n",date)
    print(100*"-")

# Enregistrer le fichier
wb.save("ventes1.xlsx")
```
